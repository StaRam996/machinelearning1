## 01 - 1 pt
- Dado un array de elementos, escribir una función, que reciba dicho array como parámetro de entrada y devuelva el elemento repetido. Usando plain Python, sin ninguna librería.

```python
array_src = [4, 8, 0, 3, 6, 2, 1, 5, 3, 9, 7]
expected_rep_element = 3
rep_element = repeated_element(array_src) # This calls your function repeated_element
assert(expected_rep_element == rep_element)  # This will fail if the element is not found correctly
```

```python
# Your solution HERE !!!!
def repeated_element(arr):
    vistos = set()
    for i in arr:
        if i in vistos:
            return i
        vistos.add(i)
    return None

```

```python

languages = ['python', 'php', 'java', 'javascript', 'c++', 'sql']
expected_output = pd.Series(['PythoN', 'PhP', 'JavA', 'JavascripT', 'C++', 'SqL'])
output = first_last_uppercase(languages)  # This calls your function first_last_uppercase
assert(expected_output.equals(output))  # This will fail if the two lists are differen
```

```python
# Your solution HERE !!!!
def first_last_uppercase(lista):
    resultado = []
    for palabra in lista:
        if len(palabra) == 1:
            nueva = palabra.upper()
        else:
            nueva = palabra[0].upper() + palabra[1:-1] + palabra[-1].upper()
        resultado.append(nueva)
    return pd.Series(resultado)
    resultado = first_last_uppercase(['python', 'php', 'java'])
print(resultado)
```

- Crear un generador infinito de números de Fibonacci.

- Crear una lista de meses de Enero a Diciembre (eje X)
- Generar datos aleatorios para el eje Y, hacerlo 3 veces distintas (Y0, Y1 e Y2)
- Representar las 3 secuencias de datos aleatorios en una sola figura usando matplotlib
- Añadir una leyenda para poder identificar cada secuencia
- Nombrar las 3 secuencias de la siguiente manera ("Facebook", "Twitter", "Instagram")
- Añadir un título y nombres a los dos ejes
- Añadir una anotación (texto y flecha) a uno de los gráficos

```python
# Your solution HERE !!!!
import matplotlib.pyplot as plt
import random

# este codigo es de chat
# Crear lista de meses (eje X)
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
         "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

# Generar datos aleatorios para cada red social
facebook = [random.randint(20, 100) for _ in meses]
twitter = [random.randint(20, 100) for _ in meses]
instagram = [random.randint(20, 100) for _ in meses]

# Crear la figura
plt.figure(figsize=(12, 6))

# Representar las 3 secuencias en un mismo gráfico
plt.plot(meses, facebook, label="Facebook", marker='o')
plt.plot(meses, twitter, label="Twitter", marker='s')
plt.plot(meses, instagram, label="Instagram", marker='^')

# Añadir leyenda
plt.legend()

# Título y nombres de ejes
plt.title("Seguidores por Mes en Redes Sociales")
plt.xlabel("Meses del Año")
plt.ylabel("Número de Seguidores (en miles)")

# Añadir una anotación (por ejemplo, punto máximo de Instagram)
max_instagram = max(instagram)
mes_max = meses[instagram.index(max_instagram)]
plt.annotate(f"Pico en {mes_max}",
             xy=(mes_max, max_instagram),
             xytext=(mes_max, max_instagram + 10),
             arrowprops=dict(facecolor='red', shrink=0.05),
             ha='center', color='red')

# Mostrar la gráfica
plt.grid(True)
plt.tight_layout()
plt.show()
```
